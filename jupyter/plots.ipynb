```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../lambda_0.5_scaling_enabled.csv')  
df
newdf = pd.read_csv('../lambda_0.5_scaling_disabled.csv')
```

```python
a = df[['time','utilisation']]
a = a.mask(a.eq('None')).dropna()
a['utilisation'] = a['utilisation'].astype(float)
plt.plot(a[["time"]], a[["utilisation"]], label='auto-scaling')

a = newdf[['time','utilisation']]
a = a.mask(a.eq('None')).dropna()
a['utilisation'] = a['utilisation'].astype(float)
plt.plot(a[["time"]], a[["utilisation"]], label='no auto-scaling')

plt.title("Utilisation")
plt.legend(loc="upper left")
plt.show()
```

```python
c = df[['time','arrival_rate']]
c = c.mask(c.eq('None')).dropna()
c['arrival_rate'] = c['arrival_rate'].astype(float)
plt.plot(c[["time"]], c[["arrival_rate"]], label='arrival rate')
plt.title("Arrival rate")
plt.legend(loc="upper left")
plt.show()
```

```python
b = df[['time','num_nodes','estimated_nodes_num']]
plt.plot(b[["time"]], b[["num_nodes"]],label="number of nodes")
# plt.plot(b[["time"]], b[["estimated_nodes_num"]],label="estimated number of nodes")
plt.title("Number of nodes")
plt.legend(loc="upper left")
plt.show()
```

```python
b = df[['time','total_average_response_time']]
b = b.mask(b.eq('None')).dropna()
b['total_average_response_time'] = b['total_average_response_time'].astype(float)
plt.plot(b[["time"]], b[["total_average_response_time"]], label='auto-scaling enabled')

b_new = newdf[['time','total_average_response_time']]
b_new = b_new.mask(b_new.eq('None')).dropna()
b_new['total_average_response_time'] = b_new['total_average_response_time'].astype(float)
plt.plot(b_new[["time"]], b_new[["total_average_response_time"]],label='no auto-scaling')

plt.title("Response Time")
plt.legend(loc="upper left")
plt.show()
```
