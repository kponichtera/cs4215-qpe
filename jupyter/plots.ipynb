```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_05_enabled = pd.read_csv('../data/lambda_0.5_scaling_enabled.csv')
df_05_disabled = pd.read_csv('../data/lambda_0.5_scaling_disabled.csv')

df_10_enabled = pd.read_csv('../data/lambda_1.0_scaling_enabled.csv')
df_10_disabled = pd.read_csv('../data/lambda_1.0_scaling_disabled.csv')

df_15_enabled = pd.read_csv('../data/lambda_1.5_scaling_enabled.csv')
df_15_disabled = pd.read_csv('../data/lambda_1.5_scaling_disabled.csv')
```

```python
fig, axs = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(20, 6))

enabled_05 = df_05_enabled[['time', 'utilisation']]
enabled_05 = enabled_05.mask(enabled_05.eq('None')).dropna()
enabled_05['utilisation'] = enabled_05['utilisation'].astype(float)
axs[0].plot(enabled_05[["time"]], enabled_05[["utilisation"]], label='auto-scaling enabled')

disabled_05 = df_05_disabled[['time', 'utilisation']]
disabled_05 = disabled_05.mask(disabled_05.eq('None')).dropna()
disabled_05['utilisation'] = disabled_05['utilisation'].astype(float)
axs[0].plot(disabled_05[["time"]], disabled_05[["utilisation"]], label='auto-scaling disabled')

axs[0].set_title("$\lambda=0.5$")
axs[0].legend(loc="upper left")

enabled_10 = df_10_enabled[['time', 'utilisation']]
enabled_10 = enabled_10.mask(enabled_10.eq('None')).dropna()
enabled_10['utilisation'] = enabled_10['utilisation'].astype(float)
axs[1].plot(enabled_10[["time"]], enabled_10[["utilisation"]], label='auto-scaling enabled')

disabled_10 = df_10_disabled[['time', 'utilisation']]
disabled_10 = disabled_10.mask(disabled_10.eq('None')).dropna()
disabled_10['utilisation'] = disabled_10['utilisation'].astype(float)
axs[1].plot(disabled_10[["time"]], disabled_10[["utilisation"]], label='auto-scaling disabled')

axs[1].set_title("$\lambda=1.0$")
axs[1].legend(loc="upper left")

enabled_15 = df_15_enabled[['time', 'utilisation']]
enabled_15 = enabled_15.mask(enabled_15.eq('None')).dropna()
enabled_15['utilisation'] = enabled_15['utilisation'].astype(float)
axs[2].plot(enabled_15[["time"]], enabled_15[["utilisation"]], label='auto-scaling enabled')

disabled_15 = df_15_disabled[['time', 'utilisation']]
disabled_15 = disabled_15.mask(disabled_15.eq('None')).dropna()
disabled_15['utilisation'] = disabled_15['utilisation'].astype(float)
axs[2].plot(disabled_15[["time"]], disabled_15[["utilisation"]], label='auto-scaling disabled')

axs[2].set_title("$\lambda=1.5$")
axs[2].legend(loc="upper left")

fig.tight_layout()

plt.savefig('utilisation.pdf')
```

```python
arrival_05 = df_05_enabled[['time', 'arrival_rate']]
arrival_05 = arrival_05.mask(arrival_05.eq('None')).dropna()
arrival_05['arrival_rate'] = arrival_05['arrival_rate'].astype(float)

num_nodes_05 = df_05_enabled[['time', 'num_nodes']]

arrival_10 = df_10_enabled[['time', 'arrival_rate']]
arrival_10 = arrival_10.mask(arrival_10.eq('None')).dropna()
arrival_10['arrival_rate'] = arrival_10['arrival_rate'].astype(float)

num_nodes_10 = df_10_enabled[['time', 'num_nodes']]

arrival_15 = df_15_enabled[['time', 'arrival_rate']]
arrival_15 = arrival_15.mask(arrival_15.eq('None')).dropna()
arrival_15['arrival_rate'] = arrival_15['arrival_rate'].astype(float)

num_nodes_15 = df_15_enabled[['time', 'num_nodes']]

fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(20, 6))

axs[0, 0].plot(arrival_05[["time"]], arrival_05[["arrival_rate"]], label='arrival rate')
axs[0, 0].set_title("$\lambda=0.5$")
axs[0, 0].legend(loc="upper left")
axs[0, 0].sharex(axs[0, 1])
axs[0, 0].sharey(axs[0, 1])

axs[0, 1].plot(arrival_10[["time"]], arrival_10[["arrival_rate"]], label='arrival rate')
axs[0, 1].set_title("$\lambda=1.0$")
axs[0, 1].legend(loc="upper left")

axs[0, 2].plot(arrival_15[["time"]], arrival_15[["arrival_rate"]], label='arrival rate')
axs[0, 2].set_title("$\lambda=1.5$")
axs[0, 2].legend(loc="upper left")
axs[0, 2].sharex(axs[0, 1])
axs[0, 2].sharey(axs[0, 1])

axs[1, 0].plot(num_nodes_05[["time"]], num_nodes_05[["num_nodes"]], label="number of nodes")
axs[1, 0].legend(loc="upper left")

axs[1, 1].plot(num_nodes_10[["time"]], num_nodes_10[["num_nodes"]], label="number of nodes")
axs[1, 1].legend(loc="upper left")
axs[1, 1].sharey(axs[1, 0])
axs[1, 1].sharex(axs[1, 0])

axs[1, 2].plot(num_nodes_15[["time"]], num_nodes_15[["num_nodes"]], label="number of nodes")
axs[1, 2].legend(loc="upper left")
axs[1, 2].sharey(axs[1, 0])
axs[1, 1].sharex(axs[1, 0])

fig.tight_layout()

plt.savefig('arrival_rate.pdf')
```
