```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('input.csv')  
df
```

```python
plt.plot(df[["time"]], df[["num_nodes"]])
plt.plot(df[["time"]], df[["estimated_nodes_num"]])
plt.show()
```

```python
a = df[['time','total_utilisation']]
a = a.mask(a.eq('None')).dropna()
a['total_utilisation'] = a['total_utilisation'].astype(float)
plt.plot(a[["time"]], a[["total_utilisation"]])
plt.show()
```

```python
b = df[['time','total_average_response_time']]
b = b.mask(b.eq('None')).dropna()
b['total_average_response_time'] = b['total_average_response_time'].astype(float)
plt.plot(b[["time"]], b[["total_average_response_time"]])
plt.show()
```
